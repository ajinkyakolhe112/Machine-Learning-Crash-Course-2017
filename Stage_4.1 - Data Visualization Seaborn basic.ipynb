```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

## Univariate Distributions

```python
x = np.random.normal(size=100)
```

```python
sns.distplot(x,rug=True)
```

```python
plt.hist(x);
```

```python
sns.distplot(x, kde=False, rug=True,bins=20);
```

## Bivariate Distributions

```python
import random

df = pd.DataFrame()

df['x'] = random.sample(range(1, 1000),100)
df['y'] = random.sample(range(1, 1000), 100)
```

```python
df.head()
```

```python
sns.jointplot(x="x",y="y",data=df,kind='scatter')#hex , kde, scatter, reg
```

```python
sns.regplot(x="x",y="y",data=df,fit_reg=True)
```

```python
sns.lmplot(x="x",y="y",data=df)
```
